```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load Iris dataset
iris = load_iris()
X = iris.data

# Clustering using K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)  # Suppressing FutureWarning
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualize the clusters
plt.scatter(X[labels == 0, 0], X[labels == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(X[labels == 2, 0], X[labels == 2, 1], s=50, c='green', label='Cluster 3')
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='black', label='Centroids')
plt.title('K-Means Clustering of Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

Titik-titik data pada plot menunjukkan sampel-sampel dari dataset Iris.
Terdapat tiga warna yang mewakili tiga kluster yang berbeda yang dihasilkan oleh algoritma K-Means. Kluster-kluster ini diidentifikasi oleh warna merah, biru, dan hijau.
Pusat-pusat kluster (centroids) ditandai dengan simbol berwarna hitam yang lebih besar. Pusat-pusat kluster ini adalah titik-titik tengah dari kluster masing-masing.
Setiap kluster memiliki titik-titik data yang berdekatan dengannya, menunjukkan bahwa sampel-sampel tersebut memiliki karakteristik yang serupa dan tergolong dalam kluster yang sama.
Plot menampilkan hubungan antara panjang sepal (Sepal Length) di sumbu x dan lebar sepal (Sepal Width) di sumbu y. Dengan demikian, kita dapat melihat bagaimana algoritma K-Means membagi sampel-sampel berdasarkan kedua fitur ini menjadi tiga kelompok yang berbeda.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler

# Load dataset
iris = pd.read_csv("../input/iris-flower-dataset/IRIS.csv")

# Understanding the data
print(iris.info())
print(iris.head(10))

# Frequency distribution of species
iris_outcome = pd.crosstab(index=iris["species"], columns="count")
print(iris_outcome)

# Filter data by species
iris_setosa = iris.loc[iris["species"] == "Iris-setosa"]
iris_virginica = iris.loc[iris["species"] == "Iris-virginica"]
iris_versicolor = iris.loc[iris["species"] == "Iris-versicolor"]

# Feature scaling
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(iris.iloc[:, [0, 1, 2, 3]])

# Determine optimal number of clusters using silhouette score
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x_scaled)
    silhouette_scores.append(silhouette_score(x_scaled, kmeans.labels_))

# Plot silhouette scores
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal Number of Clusters')
plt.show()

# Based on silhouette score, let's choose the optimal number of clusters
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + 2
print("Optimal number of clusters:", optimal_num_clusters)

# Perform KMeans clustering with optimal number of clusters
kmeans = KMeans(n_clusters=optimal_num_clusters)
kmeans.fit(x_scaled)
iris['cluster'] = kmeans.labels_

# Visualize the clusters
sns.pairplot(iris, hue='cluster')
plt.show()
```

# K-Means

K-Means adalah algoritma berbasis centroid, atau algoritma berbasis jarak, di mana kita menghitung jarak untuk menetapkan sebuah titik ke dalam sebuah klaster. Dalam K-Means, setiap klaster terkait dengan sebuah centroid.

# Cara Implementasi K-Means Clustering:

1. Pilih jumlah klaster k.
2. Pilih k titik acak dari data sebagai centroid.
3. Tentukan semua titik ke centroid klaster terdekat.
4. Hitung ulang centroid dari klaster yang baru terbentuk.
5. Ulangi langkah 3 dan 4.

```python
#Menemukan jumlah cluster yang optimal untuk klasifikasi k-means
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

# Menggunakan metode elbow untuk menentukan jumlah cluster yang optimal untuk k-means clustering

```python
plt.plot(range(1, 11), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Jumlah Kuadrat Jarak Dalam Klaster') #dalam jumlah cluster kotak
plt.show()
```
